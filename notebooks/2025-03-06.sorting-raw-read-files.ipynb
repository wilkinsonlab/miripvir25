```python
import pandas as pd
```

```python
raw_seqs = pd.read_csv("../data/raw_seqs.index.csv", header=None, names=['raw'])
raw_seqs['library'] = raw_seqs['raw'].apply(lambda x: x.split('_')[0])
raw_seqs['is_canonical'] = raw_seqs['raw'].apply(lambda x: x.find('read') != -1)
raw_seqs = raw_seqs.query('is_canonical == True').copy()
raw_seqs['primers'] = raw_seqs['raw'].apply(lambda x: x.split('_')[2])
raw_seqs['read'] = raw_seqs['raw'].apply(lambda x: x.split('_')[3].replace('.fastq.gz', ''))
raw_seqs
```

```python
raw_seqs.pivot(index=['library', 'primers'], values='raw', columns='read').reset_index().rename(columns={'library': 'library_name', 'read1': 'input_read1', 'read2': 'input_read2', }).to_csv("../scratch/raw_seqs_bylibrary.csv")
```
