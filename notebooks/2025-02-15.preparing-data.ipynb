# 15/02/2025  - Preparing data

```python
import pandas as pd
```

```python
reference = pd.read_csv("../data/sanchis2021.sup_file_1.csv", sep=";")#[['Taxid', 'Biosample', 'Representative species', 'PAB-phyto', 'PAB-symb']]
```

```python
reference
```

```python
pathogens = reference.query('patho == "P"').drop_duplicates(subset='TaxId')
pathogens
```

```python
with open("../scratch/pathogen_taxid.txt", "w") as f:
    f.write('\n'.join(pathogens['TaxId'].astype(str).values.tolist()))
```

```python
symbionts = reference.query('symb == "S"').drop_duplicates(subset='TaxId')
symbionts
```

```python
with open("../scratch/symbionts_taxid.txt", "w") as f:
    f.write('\n'.join(symbionts['TaxId'].astype(str).values.tolist()))
```
